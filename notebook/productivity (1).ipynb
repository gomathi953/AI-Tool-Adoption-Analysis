Phase 5: Hypothesis Testing

T-Test

Null Hypothesis:
Training has no impact on productivity.

```python
from scipy.stats import ttest_ind

high_training = df[df["Training_Hours"] > 15]["Productivity_Score"]

low_training = df[df["Training_Hours"] <= 15]["Productivity_Score"]

t_stat, p_value = ttest_ind(
    high_training,
    low_training
)

print("T Statistic:", t_stat)
print("P Value:", p_value)
```

Interpretation

```python
if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

Phase 6: Organization Segmentation

Create Productivity Categories

```python
conditions = [
    df["Productivity_Score"] >= 85,
    df["Productivity_Score"] >= 70,
    df["Productivity_Score"] < 70
]

labels = [
    "High Productivity",
    "Medium Productivity",
    "Low Productivity"
]

df["Segment"] = np.select(
    conditions,
    labels,
    default="Low Productivity"
)

df["Segment"].value_counts()
```

Visualize Segments

```python
sns.countplot(
    x="Segment",
    data=df
)

plt.show()
```
